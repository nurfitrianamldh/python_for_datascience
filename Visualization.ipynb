## Line Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a random number
random = []
for _ in range(50):
  random.append(np.random.randint(1, 100))

# Create a dataframe format and save a random number inside
df = pd.DataFrame(random, columns = ['tes1'])
df
# df.head(20)
# df.tail()
```

```python
# Create a single line chart
plt.figure(figsize = (10, 5)) # ukuran canvas panjang=10 tinggi=5
plt.plot(df['tes1'], marker = 'P') # definiskan dataframe
plt.title('Single Line Chart with Random Numbers') # judul chart
plt.xticks(df.index) # skala X berdasarkan index baris
plt.xlabel('Row') # memberi label pada X
plt.ylabel('Random Numbers') # memberi label pada Y
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

# data_df
plt.figure(figsize = (20, 5))
plt.plot(xpoints, ypoints)
plt.title('Single Line Chart with Random Numbers')
plt.xlabel('X Points')
plt.ylabel('Y Points')
```

```python
# Create an another random number
random2 = []
for _ in range(50):
  random2.append(np.random.randint(1, 100))
df['tes 2'] = random2

# Create a multiple line chart
plt.figure(figsize = (10, 5))
plt.plot(df['tes1'], label = 'Random')
plt.plot(df['tes 2'], label = 'Random 2', color = 'brown')
plt.title('Multiple Line Chart with Random Numbers')
plt.xlabel('Row')
plt.ylabel('Random Numbers')
plt.legend(loc = 'upper left') # Must set label first in plt.plot()
plt.show()
```

## Bar Chart

```python
# Data preparation
x_value = ['A', 'B', 'C', 'D', 'E']
y_value = []
for _ in x_value:
  y_value.append(np.random.randint(10, 50))
df = pd.DataFrame({'Nilai X' : x_value, 'Nilai Y' : y_value})


# Create a bar chart
plt.figure(figsize = (15, 8))
plt.bar(df['Nilai X'], df['Nilai Y'])
plt.title('Frequency of Each Alphabet')
plt.xticks(df['Nilai X'])
plt.yticks(np.arange(0, 51, step = 10))
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.show()
```

```python
# Horizontal Bar Chart
plt.figure(figsize = (20, 8))
plt.barh(df['Nilai X'], df['Nilai Y'], color='r', height = 0.5)
plt.title('Frequency of Each Alphabet')
plt.yticks(df['Nilai X'])
plt.xticks(np.arange(0, 51, step = 10))
plt.xlabel('Nilai Y')
plt.ylabel('Nilai X')
plt.show()
```

## Pie Chart

```python
# Data preparation
perc = [20, 15, 40, 25]
mylabel = ['A', 'B', 'C', 'D']
# Create a pie chart
plt.figure(figsize = (10, 5))
plt.pie(perc, labels = mylabel, autopct = '%1.1f%%')
plt.title('Percentage of Alphabet')
plt.show()
```

```python
# Pie chart explode
nilai = [20, 15, 40, 25]
pielabel = ["Apples", "Bananas", "Cherries", "Dates"]
nilai_explode = [0, 0, 0.1, 0]

# Create a pie chart
plt.figure(figsize = (10, 5))
plt.pie(nilai, labels = pielabel, autopct = '%1.1f%%', explode = nilai_explode)
plt.title('Percentage of Alphabet')
plt.show()
```

```python
# Pie chart colors
nilai = [20, 15, 40, 25]
pielabel = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["orange", "brown", "b", "#4CAF50"]

# Create a pie chart
plt.figure(figsize = (10, 5))
plt.pie(nilai, labels = pielabel, autopct = '%1.1f%%', colors = mycolors)
plt.title('Percentage of Alphabet')
plt.show()
```

## Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a random number
nilai_x = []
for _ in range(50):
  nilai_x.append(np.random.randint(1, 100))
# Create a dataframe format and save a random number inside
df = pd.DataFrame(nilai_x, columns = ['Nilai X'])

nilai_y = []
for _ in range(50):
  nilai_y.append(np.random.randint(1, 100))
df['Nilai Y'] = nilai_y

#Scatter Plot
plt.figure(figsize = (20, 5))
plt.scatter(df['Nilai X'], df['Nilai Y'])
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
# plt.grid()
plt.show()
```

```python
#Scatter Plot one color
plt.figure(figsize = (20, 5))
plt.scatter(df['Nilai X'], df['Nilai Y'], color = '#88c999')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.grid()
plt.show()
```

```python
#Scatter Plot multiple color
mycolor = []
for _ in range(50):
  mycolor.append(np.random.randint(1, 100))

plt.figure(figsize = (20, 5))
plt.scatter(df['Nilai X'], df['Nilai Y'],  c=mycolor)
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.grid()
plt.colorbar()
plt.show()
```

## Intro

```python
import plotly.express as px


# Creating the Figure instance
fig = px.line(x=[1, 2, 3], y=[1, 2, 3])

# showing the plot
fig.show()
```

```python
import plotly.express as px
fig = px.scatter(x=[0, 1, 2, 3, 4], y=[0, 1, 4, 9, 16])
fig.show()
```

```python
import plotly
import plotly.graph_objects as go

x = np.random.randint(low=1, high=50, size=150)*0.1

# generating 150 random integers
# from 1 to 50
y = np.random.randint(low=1, high=50, size=150)*0.1

# plotting scatter plot
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers'))

fig.show()
```

```python
import plotly.express as px
data_canada = px.data.gapminder().query("country == 'Canada'")
fig = px.bar(data_canada, x='year', y='pop')
fig.show()
```

```python
countries=['India', 'canada',
           'Australia','Brazil',
           'Mexico','Russia',
           'Germany','Switzerland',
           'Texas']

# plotting corresponding y for each
# country in x
fig = go.Figure([go.Bar(x=countries,
                        y=[80,70,60,50,
                           40,50,60,70,80])])

fig.show()
```

```python
values = [4500, 2500, 1053, 500,
          3200, 1500, 1253, 600, 3500]

# plotting pie chart
fig = go.Figure(data=[go.Pie(labels=countries,
                      values=values)])

fig.show()
```
